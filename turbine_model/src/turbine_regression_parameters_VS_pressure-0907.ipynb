```python
import os
import sys
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from turbine_regression import get_origin_data
from turbine_regression import data_preprocessing
from turbine_regression import get_result_data
```

## 压力平均数据

```python
origin_data = get_origin_data(period = 3)
origin_data = origin_data[[
    "eturb_m1_steam_pressure_in", "eturb_m2_steam_pressure_in", "bturb_m1_steam_pressure_in",
    "eturb_m1_steam_temperature_in", "eturb_m2_steam_temperature_in", "bturb_m1_steam_temperature_in",
    "eturb_m1_steam_pressure_side", "eturb_m2_steam_pressure_side", "bturb_m1_steam_pressure_out",
    "eturb_m1_steam_temperature_side", "eturb_m2_steam_temperature_side", "eturb_m1_steam_temperature_out",
    "eturb_m1_steam_pressure_out", "eturb_m2_steam_pressure_out", 
    "eturb_m1_steam_temperature_out", "eturb_m2_steam_temperature_out",
]]
origin_df = data_preprocessing(origin_data, method = "mean", is_export_csv = 0).reset_index(drop = True)
origin_df
```

## 回归结果

```python
result_df = get_result_data(period = 3)
result_df_origin = result_df[["eturb_m1_origin_k1", "eturb_m1_origin_k2", "eturb_m2_origin_k1", "bturb_m1_origin_k1"]].reset_index(drop = True)
result_df_delta = result_df[["eturb_m1_delta_k1", "eturb_m1_delta_k2", "eturb_m2_delta_k1", "bturb_m1_delta_k1"]].reset_index(drop = True)
result_df_base = result_df[["eturb_m1_base_k1", "eturb_m1_base_k2", "eturb_m2_base_k1", "bturb_m1_base_k1"]].reset_index(drop = True)
result_df_origin
```

```python
# Origin k1、k2
df_origin = pd.concat([result_df_origin, origin_df], axis = 1, sort = False)
df_origin.head()
```

```python
# Delta k1、k2
df_delta = pd.concat([result_df_delta, origin_df], axis = 1, sort = False)
df_delta.head()
```

```python
# Base k1、 k2
df_base = pd.concat([result_df_base, origin_df], axis = 1, sort = False)
df_base.head()
```
