```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

root = "e:/dev/data-analysis/goodness_of_fit_score"
```

```python
def scatter_plot(df, xlabel, *ylabels, imgpath = None):
    """
    matlab plot style
    """
    fig, axs = plt.subplots(figsize = (10, 7), sharey = True, tight_layout = True)
    for ylabel in ylabels:
        plt.scatter(df[xlabel], df[ylabel], marker = "o", label = ylabel)
        plt.title("%s VS %s" % (xlabel, ylabel))
        plt.xlabel("%s" % xlabel)
        # plt.ylabel("%s" % ylabel)
        plt.legend()
        if imgpath:
            plt.savefig(imgpath)
        plt.show()
```

```python
data = pd.read_csv("e:/dev/data-analysis/goodness_of_fit_score/goodness_of_fit_score.csv", header = 0, index_col = None)
data
```
